```python
# heat maps
big_columns = ["duration_ms", "loudness", "tempo"]

for col in big_columns:
    # Excluding this one column
    no_col_df = songs_df.drop(columns=[col]).copy()
    no_col_distances = pd.DataFrame(
        pairwise_distances(no_col_df),
        index=no_col_df.index,
        columns=no_col_df.index
    )
    
    # Including all columns (original)
    with_col_distances = pd.DataFrame(
        pairwise_distances(songs_df),
        index=songs_df.index,
        columns=songs_df.index
    )
    
    # Plot heatmaps
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    sns.heatmap(no_col_distances, ax=ax1, linewidths=.2, cmap="viridis")
    sns.heatmap(with_col_distances, ax=ax2, linewidths=.2, cmap="viridis")
    
    ax1.set_title(f"Excluding {col}")
    ax2.set_title(f"Including {col}")
    
    plt.show()
```

```python
# duration_ms → Heatmap structure changes significantly.
# durations are in the hundreds of thousands of ms (100,000–400,000), way larger than 0–1 features like danceability.

# loudness → Range is about -33 to -3 → still only ~30 units wide, which is closer to other features.
# So it doesn’t dominate like duration.

# tempo → Moderate range (60–160), but still not huge.
# Affects distances a little, but not enough to overwhelm the others.

```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(data_quantile)

print("Original shape:", data_quantile.shape)
print("Reduced shape:", X_pca.shape)
print("Total explained variance:", pca.explained_variance_ratio_.sum())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

inertias = []
silhouettes = []
K_range = range(2, 6)  # try 2 to 5 clusters for 10 songs

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_pca)
    inertias.append(model.inertia_)
    silhouettes.append(silhouette_score(X_pca, labels))

# Elbow plot
plt.plot(K_range, inertias, marker="o")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()

# Silhouette plot
plt.plot(K_range, silhouettes, marker="o")
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Analysis")
plt.show()
```

```python
# Radar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# If data_quantile is a NumPy array, wrap it into a DataFrame with column names
# Replace these names with your actual feature names
feature_names = songs_df_sorted.select_dtypes(include=[np.number]).drop(
    columns=['cluster','tsne0','tsne1'], errors='ignore'
).columns

scaled_df = pd.DataFrame(data_quantile, columns=feature_names)

# Add cluster labels from your model
scaled_df['cluster'] = model.labels_

# Compute per-cluster averages
cluster_means = scaled_df.groupby('cluster')[feature_names].mean()

# --- Radar chart ---
N = len(feature_names)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # complete the loop

fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))

# Plot each cluster's mean profile
for cl, row in cluster_means.iterrows():
    values = row.tolist()
    values += values[:1]  # close the loop
    ax.plot(angles, values, linewidth=2, label=f'Cluster {cl}')
    ax.fill(angles, values, alpha=0.25)

# Add feature labels around the circle
ax.set_xticks(angles[:-1])
ax.set_xticklabels(feature_names, fontsize=10)

# Scaled data is already in [0, 1]
ax.set_ylim(0, 1)

ax.set_title("Radar Chart of Scaled Song Features by Cluster", size=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.15, 1.15))
plt.show()
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

def sweep_dbscan(X, min_samples, eps_center, width=0.1, steps=7):
    eps_vals = np.round(np.linspace(eps_center*(1-width), eps_center*(1+width), steps), 3)
    rows = []
    for e in eps_vals:
        labels = DBSCAN(eps=e, min_samples=min_samples).fit_predict(X)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        noise = (labels == -1).mean()
        sil = None
        if n_clusters > 1 and noise < 1.0:
            sil = silhouette_score(X, labels)
        rows.append((e, n_clusters, noise, sil))
    return rows

print("min_samples=2")
for r in sweep_dbscan(X_pca, 2, 0.92, width=0.05):
    print(f"eps={r[0]:.3f} -> clusters={r[1]}, noise={r[2]:.0%}, silhouette={r[3]}")

print("\nmin_samples=3")
for r in sweep_dbscan(X_pca, 3, 1.0, width=0.05):
    print(f"eps={r[0]:.3f} -> clusters={r[1]}, noise={r[2]:.0%}, silhouette={r[3]}")
```
